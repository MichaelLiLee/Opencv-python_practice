```python
#引入opencv並顯示版本
import cv2
print(cv2.__version__)
```

---
### 1.1.讀取圖片(Read an image)

使用`cv2.imread()`讀取圖片，圖片會儲存成一個 NumPy 的陣列，我們可以用 type 檢查一下。


```python
img = cv2.imread('cat.jpg')
type(img)
```

---
### 1.2.顯示圖片(Display an image)
#### 方法1:

使用`cv2.imshow()`顯示圖片

接著使用`cv2.waitkey(0)`&`cv2.destoryAllWindow()`來觸發當使用者按下任意按鍵時，關閉視窗的功能

**注意 : 單獨使用`cv.imshow()`而沒有同時使用`cv2.waitkey(0)`&`cv2.destoryAllWindow()`時會發生錯誤**


```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

如果想要將圖片顯示在大小可以調整的視窗時，方法如下:

使用`cv2.nameWindow()`創建一個可調整視窗大小的視窗

接著使用`cv2.imshow()`顯示圖片

最後使用`cv2.waitkey(0)`&`cv2.destoryAllWindow()`來觸發當使用者按下任意按鍵時，關閉視窗的功能

**注意 : 單獨使用`cv.imshow()`而沒有同時使用`cv2.waitkey(0)`&`cv2.destoryAllWindow()`時會發生錯誤**


```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 方法2:

使用matplotlib中的`plt.imshow()`

**注意:由於opencv讀取的圖片是以BGR格式儲存，因此需要轉換成RGB格式**`img_rgb = img_bgr[:,:,::-1]`


```python
#引入numpy & matplotlib.pyplot & opencv
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img_bgr = cv2.imread('cat.jpg')

#將圖片從BGR格式轉換成RGB格式
img_rgb = img_bgr[:,:,::-1]

plt.imshow(img_rgb)
```

### 1.3.儲存圖片(Write an image)

目的 : 將讀取的圖片儲存成圖檔

使用function : `cv2.imwrite()`


```python
cv2.imwrite('cat2.jpg', img)
```
